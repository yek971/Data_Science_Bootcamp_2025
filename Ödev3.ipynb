```python
tp.groupby(["day", "time"])["total_bill"].agg(["sum", "min", "max", "mean"])
```

```python
tp[(tp["sex"] == "Female") & (tp["time"] == "Lunch")] \
  .groupby("day")[["total_bill", "tip"]] \
  .agg(["sum", "min", "max", "mean"])
```

```python
tp["total_bill_tip_sum"] = tp["total_bill"] + tp["tip"]
tp.head()
```
